```python
import pandas as pd
import matplotlib.pyplot as plt



category_counts = df['traffic_control_device'].value_counts()

category_counts.plot(kind='bar', color='skyblue', figsize=(10, 6), title='Traffic Control Device Distribution', xlabel='Traffic Control Device', ylabel='Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['injuries_total'].plot(kind='hist',color='green', figsize=(10, 6), title='Injuries Total Distribution', xlabel='Injuries Total', ylabel='Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['traffic_control_device'].value_counts().plot(kind='bar', color='skyblue', figsize=(10, 6), title='Traffic Control Device Distribution', xlabel='Traffic Control Device', ylabel='Count', rot=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['crash_date'] = pd.to_datetime(df['crash_date'])
df.groupby('crash_date')['injuries_total'].sum().plot(kind='line', figsize=(10, 6), title='Total Injuries Over Time', xlabel='Crash Date', ylabel='Total Injuries')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.scatter(df['injuries_total'], df['num_units'], color='green'); plt.xlabel('Injuries Total'); plt.ylabel('Number of Units'); plt.show()
```
